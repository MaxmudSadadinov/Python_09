# График №1

````
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````
data = pd.read_csv('/content/spotify-2023_fixed.csv')

data
````

**Разделяем значения на группы.**
Значение столбика разделяются на группы. Для них создается новый столбик. Перед каждым значением возврашается его группа

````
data.loc[data['released_year'] <= 1990, 'star in year'] = 'исп. 80ых'
data.loc[(data['released_year'] > 1990) & (data['released_year'] <= 2000), 'star in year'] = 'исп. 90ых'
data.loc[(data['released_year'] > 2000) & (data['released_year'] <= 2010), 'star in year'] = 'исп. 2000ых'
data.loc[(data['released_year'] > 2010) & (data['released_year'] <= 2020), 'star in year'] = 'исп. 2010ых'
data.loc[data['released_year'] > 2020, 'star in year'] = 'Современные исп.'
data
````

Загооловок

````
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'sans-serif'}
lable_dict = {'fontsize': '14', 'color': '#808080', 'family': 'sans-serif'}
````

````
sns.set_theme(style="whitegrid")
sns.relplot(data=data, x='star in year', y='energy_%', kind='line', height=5, aspect=1.8)
plt.title("Топ по энергичности", fontdict=title_dict)
plt.xlabel("Музыканты", fontdict=lable_dict)
plt.ylabel("Энергичность их музыки", fontdict=lable_dict)
plt.show()
````

# Вывод к первому графику
